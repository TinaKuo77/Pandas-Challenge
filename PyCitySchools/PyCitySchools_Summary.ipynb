```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="outer", on=["school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_number_schools = len(school_data_complete["School ID"].unique())
print(f"The Total Number of Schools is {round(total_number_schools,2)}")
```

```python
#Calculate the total number of students
total_number_students = len(school_data_complete["Student ID"].unique())
print(f"The Total Number of Students is {round(total_number_students,2)}")
```

```python
#Calculate the total budget
total_budget = school_data["budget"].sum()
print(f"The Total Budget is {round(total_budget,2)}")
```

```python
#Calculate the average math score
average_math_score = student_data["math_score"].mean()
print(f"The average Math Score is {round(average_math_score,2)}")
```

```python
#Calculate the average reading score
average_reading_score = student_data["reading_score"].mean()
print(f"The average Reading Score is {round(average_reading_score,2)}")
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
student_data["#passing_math"] = student_data["math_score"] >= 70
percent_passing_math = ((student_data["#passing_math"]).mean())*100
print(f"The percentage of students with a passing math score (70 or greater) is {round(percent_passing_math,2)} % ")
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
student_data["#passing_reading"] = student_data["reading_score"] >= 70
percent_passing_reading = ((student_data["#passing_reading"]).mean())*100
print(f"The percentage of students with a passing reading score (70 or greater) is {round(percent_passing_reading,2)} % ")
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = (percent_passing_math + percent_passing_reading)/2
print(f"The percentage of students who passed math and reading is {round(overall_passing,2)} % ")
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
district_results = [{"Total Schools": total_number_schools, 
            "Total Students": total_number_students, 
            "Total Budget": total_budget, 
            "Average Math Score":  round(average_math_score,2), 
            "Average Reading Score":  round(average_reading_score,2), 
           "Passing Math (%)": round(percent_passing_math,2),
           "Passing Reading (%)": round(percent_passing_reading,2),
            "Overall Passing Rate (%)": round(overall_passing,2)}]
Summary_table = pd.DataFrame(district_results)

#Formatting
Summary_table["Total Students"] = Summary_table["Total Students"].map("{:,.0f}".format)
Summary_table["Total Budget"] = Summary_table["Total Budget"].astype(float).map("${:,.2f}".format)
Summary_table["Passing Math (%)"] = Summary_table["Passing Math (%)"].map("{:.2f}%".format)
Summary_table["Passing Reading (%)"] = Summary_table["Passing Reading (%)"].map("{:.2f}%".format)
Summary_table["Overall Passing Rate (%)"] = Summary_table["Overall Passing Rate (%)"].map("{:.2f}%".format)

Summary_table
```

```python
#Group by School Name

grouped_school_data_complete = school_data_complete.groupby(['school_name'])
school_typ = []
total_students = []
total_budgets = []
per_studnet_budgets = []
average_math = []
average_Reading = []
number_pass_math = []
percent_pass_math = []
percent_pass_reading = []
percent_overall_pass = []

school_type = grouped_school_data_complete["type"].unique().str[0]
#school_type
```

```python
# total students of each schools
total_students = grouped_school_data_complete["school_name"].count()
#total_students
```

```python
# total bugets of each schools
total_budgets = grouped_school_data_complete["budget"].unique()
#total_budgets
```

```python
# Per Student budget of each schools
per_studnet_budgets = total_budgets/total_students
#per_studnet_budgets
```

```python
#Average Math Score of each schools
average_math = grouped_school_data_complete['math_score'].mean()
#average_math
```

```python
#Average Reading Score of each schools
average_Reading = grouped_school_data_complete['reading_score'].mean()
#average_Reading
```

```python
# % Passing Math (The percentage of students that passed math.
# A table for students with a passing math score (70 or greater) and a passing reading score (70 or greater)

school_data_complete["passing_math"] = school_data_complete["math_score"] >= 70
school_data_complete["passing_reading"] = school_data_complete["reading_score"] >= 70
```

```python
#drop math_fail
number_pass_math = school_data_complete[school_data_complete['math_score']>=70].groupby('school_name')['math_score'].count()
#number_pass_math

percent_pass_math = (number_pass_math/ total_students) *100
#percent_pass_math
```

```python
# % Passing Reading (The percentage of students that passed reading.)
#drop reading_fail
number_pass_reading = school_data_complete[school_data_complete['reading_score']>=70].groupby('school_name')['reading_score'].count()

#Passing Reading percentage
percent_pass_reading = (number_pass_reading / total_students) *100
#percent_pass_reading
```

```python
# % Overall Passing (The percentage of students that passed math and reading.)
percent_overall_pass = (percent_pass_reading + percent_pass_math)/2
#percent_overall_pass
```

```python
# Create a dataframe to hold the above results

for school in grouped_school_data_complete: 
    Summary_table = pd.DataFrame( {"School Type": school_type, 
            "Total Students": total_students, 
            "Total Budget": total_budgets, 
            "Per Student Budget": per_studnet_budgets, 
            "Average Math Score":  round(average_math,2), 
            "Average Reading Score":  round(average_Reading,2), 
           "Passing Math (%)": round(percent_pass_math,2),
           "Passing Reading (%)": round(percent_pass_reading,2),
            "Overall Passing Rate (%)": round(percent_overall_pass,2)})
    
#Formatting
Summary_table["Total Students"] = Summary_table["Total Students"].map("{:,.0f}".format)
Summary_table["Total Budget"] = Summary_table["Total Budget"].astype(float).map("${:,.2f}".format)
Summary_table["Per Student Budget"] = Summary_table["Per Student Budget"].astype(float).map("${:,.2f}".format)
Summary_table["Passing Math (%)"] = Summary_table["Passing Math (%)"].map("{:.2f}%".format)
Summary_table["Passing Reading (%)"] = Summary_table["Passing Reading (%)"].map("{:.2f}%".format)
Summary_table["Overall Passing Rate (%)"] = Summary_table["Overall Passing Rate (%)"].map("{:.2f}%".format)

Summary_table
```
